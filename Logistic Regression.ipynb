```r
# LOGISTIC REGRESSION

Marketing=Marketing[,-11] #delete the call-duration variable
View(Marketing)
str(Marketing)

Models0 <-glm(Target ~ 1,  family=binomial(link="logit"), data=Marketing) #model without regressors
summary(Models0)

Models <-glm(Target ~ .,  family=binomial(link="logit"), data=Marketing) #model with regressors
summary(Models)

library(lmtest)
lr_test=lrtest(Models, Models0)
print(lr_test) #likelhood test corroborates the model with regressors

## STEPWISE TO SELECT THE BEST VARIABLES
library(MASS)
library(dplyr)
library(car)
library(caret)
 
step.model1= step(Models)
vif(step.model1)
```

```r
## TRAINING AND TEST SET

ind <- sample(2, nrow(Marketing), replace=TRUE, prob=c(0.70, 0.30))

ftable(ind) ## 28804 vs 12384

## I build the sample to fit the method
Marketing.training <- Marketing[ind==1, c(2,5,8,9,10,11,12,14,15,16,17,18,19)] 

View(Marketing.training)
nrow(Marketing.training) #2804

##Logistic Regression algorithm
Models.training0=glm(Target ~1,  family=binomial(link="logit"), data=Marketing.training)
summary(Models.training0)

Models.training=glm(Target ~.,  family=binomial(link="logit"), data=Marketing.training)
summary(Models.training)

#TRAIN SET PREDICTION
yprob <-predict(Models.training,Marketing.training,"response")
y_prev<- ifelse(yprob>0.56,'1','0')
y_oss <- Marketing.training$Target
View(cbind(y_prev,y_oss))

tab <- xtabs(~y_prev+y_oss)
tab
```

```r
#TEST SET
Marketing.test <- Marketing[ind==2, c(2,5,8,9,10,11,12,14,15,16,17,18,19)]
nrow(Marketing.test) #12384
View(Marketing.test)

##TEST-SET PREDICTION
yprob <-predict(Models.training,Marketing.test,"response")
y_prev<- ifelse(yprob>0.40,'1','0')
y_oss <- Marketing.test$Target
View(cbind(y_prev,y_oss))

tab <- xtabs(~y_prev+y_oss)
tab
```

```r
#METRICS FOR EVALUATE THE PERFORMANCE OF THE LOGISTIC MODEL

T0 = sum((y_prev==0&y_oss==0))## 0 correctly predicted
T1 = sum((y_prev==1&y_oss==1)) ## 1 correctly predicted
F0 = sum((y_prev==0&y_oss==1)) ## 1 incorrectly predicted
F1 = sum((y_prev==1&y_oss==0)) ## 0 incorrectly predicted

## ACCURACY AND SPECIFICITY
n <- nrow(Marketing.test)
accuracy=((T0+T1)/n)*100
accuracy #89.95%

specificita=(T1/(T1+F0))*100   
specificita

#SENSITIVITY AND PRECISION
sensitivita =(T0/(T0+F1))*100 #### TPR true positive rate
sensitivita
ST=((F0+F1)/n)*100 ## standard error
precisione=(T0/(T0+F0))*100 # correctly predicted positive values: 90%
precisione

##McFadden
library(pscl)
library(forecast)
pscl::pR2(Models.training)["McFadden"] #0.215945

##PseudoR2
pR2(Models.training)['r2ML'] #0.1411 Pseudo R-2

##Forecast R2
FR2=1-(logLik(Models.training)/(logLik(Models.training0)))
FR2 #0.215945 


##ROC CURVE AND AREA UNDER the CURVE (AUC)
library(pROC)
lrROC <- roc(y_oss ~ yprob,plot=TRUE,print.auc=TRUE,col="black",lwd =1.4,legacy.axes=TRUE,main="ROC Curves")
auc(lrROC) #79.4% GOOOD!
```
